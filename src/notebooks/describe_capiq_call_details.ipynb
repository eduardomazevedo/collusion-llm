```python
import pandas as pd

df = pd.read_feather('./data/transcript-detail.feather')
df.head()
```

```python
df.dtypes
```

```python
df.describe(include='all')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Extract the year from the 'mostimportantdateutc' column
df['year'] = pd.to_datetime(df['mostimportantdateutc']).dt.year

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['year'], bins=range(2004, 2026), kde=False)
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Histogram of Most Important Dates by Year')
plt.show()
```

```python
# Convert audiolengthsec to minutes
df['audiolengthmin'] = df['audiolengthsec'] / 60

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['audiolengthmin'].dropna(), bins=range(0, 301), kde=False)
plt.xlabel('Audio Length (minutes)')
plt.ylabel('Frequency')
plt.title('Histogram of Audio Length in Minutes')
plt.xlim(0, 200)
plt.show()
```
