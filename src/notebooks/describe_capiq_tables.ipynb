# Describe Capital IQ Transcripts Tables Available in WRDS

```python
import config
import wrds
import pandas as pd

conn = wrds.Connection(wrds_username=config.WRDS_USERNAME, password=config.WRDS_PASSWORD)
ciq_tables = conn.list_tables(library='ciq')
transcript_tables = [table for table in ciq_tables if 'transcript' in table]
```

## Table names and stats

```python
ciq_table_stats = []

for table in transcript_tables:
    query = f"SELECT * FROM ciq.{table} LIMIT 1"
    df = conn.raw_sql(query)
    nrow = conn.get_row_count(library='ciq', table=table)
    ncol = len(df.columns)
    size_gb = df.memory_usage(deep=True).sum() * nrow / (1024 ** 3)
    col_names = df.columns.tolist()
    
    ciq_table_stats.append({
        'name': table,
        'nrow': nrow,
        'ncol': ncol,
        'size_gb': size_gb,
        'col_names': col_names
    })

ciq_table_stats = pd.DataFrame(ciq_table_stats)
ciq_table_stats
```

## List variables

```python
for index, row in ciq_table_stats.iterrows():
    print(f"Table: {row['name']}")
    print("Columns:")
    for col in row['col_names']:
        print(f" - {col}")
    print("\n")
```
